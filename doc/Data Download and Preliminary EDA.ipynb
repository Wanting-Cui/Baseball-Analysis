```python
import pandas as pd
import numpy as np
from pybaseball import statcast
from pybaseball import pitching_stats_range

pd.set_option('display.max_columns', 100)
```

```python
# Data Download
dat_04 = statcast(start_dt='2017-04-01', end_dt='2017-04-30')
```

```python
dat_05 = statcast(start_dt='2017-05-01', end_dt='2017-05-31')
dat_06 = statcast(start_dt='2017-06-01', end_dt='2017-06-30')
dat_07 = statcast(start_dt='2017-07-01', end_dt='2017-07-31')
dat_08 = statcast(start_dt='2017-08-01', end_dt='2017-08-31')
dat_09 = statcast(start_dt='2017-09-01', end_dt='2017-09-30')
dat_10 = statcast(start_dt='2017-10-01', end_dt='2017-11-01')
```

```python
dat_2017 = pd.concat([dat_04, dat_05, dat_06, dat_07, dat_08, dat_09, dat_10])
dat_2017.to_csv("../input/dat1_2017.csv")
```

```python
dat_group = dat_2017.groupby('player_name')
pitch_num = dat_group.size().sort_values(ascending=False)
print("Top amount of pitches throw")
print(pitch_num.head(10))
print("Bottom amount of pitches throw")
print(pitch_num.tail(10))
speed = dat_group['release_speed'].agg(np.mean).sort_values(ascending=True)
print("Bottom speed")
print(speed.head(10))
print("Top speed")
print(speed.tail(10))

import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(pitch_num)
plt.savefig("../fig/Hist Number of Pitches.jpg")
# plt.show()
```
